## 1.2 复变函数

### 一些基本概念

#### 复变量与复变函数

设 $E\in \mathbb{C}$ 为复数平面 (或扩充复平面) 上的一个点集 (复数的集合), 给定一个复数 $z=x + i y$, 其取值可以是 $E$ 内的任意值, $z$ 称为复变量 (complex variable). 若对于 $E$ 内的任意一个复数 $z$, 都有一个唯一确定的复数 $w = u + i v$ 按照一定的规律与之相对应，则称在 $E$ 上定义了一个复变函数 (complex function), 简称函数, 记作 
$$w = f(z) \quad (z\in E).$$
$z$ 称为自变量 (independent variable), $E$ 称为函数的定义域 (domain of definition 或简称 domain). 

如果和每个点 $z\in E$ 对应的复数 $w$ 不止有一个 (有时甚至是无限多个), 我们便需要扩充函数的概念, 将这样的函数称为多值 (multiple-valued 或 multivalued) 函数. 
上述有且仅有一个 $w$ 值与 $z\in E$ 对应的函数则称为单值 (single-valued) 函数. 多值函数举例:
- $w = \sqrt[n]{z}$ ($n\in \mathbb{N}$), 当 $n\geqslant2$ 时有 $n$ 种可能取值; 
- $w = \ln(z) = \ln(|z|) + i\operatorname{Arg}(z) = \ln(|z|) + i\arg(z) + i\,2 k\pi$ $(k\in \mathbb{Z})$ 有无限多种取值.

将 $z$ 取遍 $E$ 内的值, 得到的 $w$ 的集合 $E'$ 称为函数 $w=f(z)$ 的值域 (range). $E'\in \mathbb{C}$ 也是复数平面 (或扩充复平面) 上的一个点集; 因此, $w$ 也是一个复变量, 称为因变量 (dependent variable). 由于因变量 $w = u + i v$ 和自变量 $z= x + i y$ 都分别有实部和虚部, 给定 $w = f(z)$ 便相当于确定了两个二元实变函数:
$$ u = u(x,y), \quad v  = v(x,y). $$
例如, $w = z^2$ 可写为
$$ u + i v = (x + i y)^2 = x^2 - y^2 + 2 i xy, $$
即对应于下列两个实变函数:
$$ u = x^2 - y^2, \quad v = 2 xy. $$ 

#### 复变函数的几何意义

复变函数是实变函数的推广, 但是它们的几何意义差别很大. 例如, 考虑一个简单的实变函数 $y = x^2$, 它可以表示为一条抛物线. 
但是, 对于复变函数 $w = z^2$ 来说, 由于 $w$ 和 $z$ 分别都是复变量, 要完全将它们的关系表达出来需要 4 维空间的几何图形, 无法在一个图形中画出. 然而, 我们可以通过映射来赋予复变函数更直观的几何意义. 我们取两个复数平面, 分别为 $z$ 平面和 $w$ 平面. 定义域 $E$ 和值域  $E'$ 分别是 $z$ 平面和 $w$ 平面上的点集, 如此, 则 $w = f(z)$ 便可看做从 $z$ 平面上的 $E$ 到 $w$ 平面上的 $E'$ 的一个映射 (mapping), 或者看做将 $z$ 平面上的 $E$ 变换成了 $w$ 平面上的 $E'$; $w$ 称为 $z$ 的像 (image), $z$ 称为 $w$ 的原像 (inverse image).


我们来看一下曲线的概念. 平面上的曲线 (curve) 可由一对参数方程来描述:
 \begin{equation}
    x=x(t), \quad y=y(t) \quad( t \in [a, b]); \label{eq:curve}
 \end{equation}
曲线上点的坐标为 $(x(t), y(t))$ $(t \in [a, b])$. 当 $x(t)$ 和 $y(t)$ 是连续的实函数时, 上述参数方程给出来的曲线是连续的. 令 $z=x+i y, z(t)=x(t)+i y(t)$, \eqref{eq:curve} 式可以表达为一个复参数方程
\begin{equation}
    z=z(t) \quad(t \in [a, b]).
\end{equation}
$z(a)$ 和 $z(b)$ 分别为曲线的起点和终点. 如果 $z(a) = z(b)$, 即起点与终点相同, 则曲线为闭曲线. 如果存在某两个参数值使得 $z(t_1) = z(t_2)$ $(t_1\neq t_2)$, 即曲线上有两个点重合, 则 $z(t_1)$ 称为曲线的重点. 没有重点的曲线称为简单曲线 (simple curve) 或约当曲线 (Jordan curve). 除起点和终点外, 没有其它重点的闭曲线称为简单闭曲线.
    
**例1:** (1) 连接 $z_1$ 和 $z_2$ 两点的直线段参数方程可写为 
\begin{equation}
z = z_1 + (z_2 - z_1) t \quad (t \in  [0,1]). \label{eq:segment}
\end{equation}
    
(2) 圆心为 $z_0$, 半径为 $\rho$ 的圆的参数方程: 这样的圆上的点满足 $|z - z_0| = \rho$, 于是, 其参数方程可写为
\begin{equation}
    z = z_0 + \rho\, e^{i\varphi} = z_0 + \rho \left(\cos\varphi + i\sin\varphi \right) \quad (\varphi\in [0, 2\pi]). \label{eq:circle}
\end{equation}

**例 2:** 考虑 $w = 1/z$ 的映射下, (1) $z$ 平面内的圆 $|z|=\rho >0$ 变成 $w$ 平面上的什么曲线? (2) $z$ 平面内的等腰直角三角形: $0\leqslant y \leqslant x, x\in [0,1]$, 求其在 $w$ 平面内的像集.

**解:** 
(1)  由 $|z| = \rho$ 可得
$$
|w|=\frac{1}{|z|}=\frac{1}{\rho}, 
$$
因此, $|z|= \rho$ 被映射为 $w$ 平面内的圆心位于原点, 半径为 $1/\rho$ 的圆.

(2) 我们分段考察 $z$ 平面内等腰三角形的三条边.
- 斜边. 令 $z = \rho e^{i\varphi}$, 则斜边可表示为 $\rho \in [0,\sqrt{2}], \varphi = \pi/4$. 于是,
$$
w=\frac{1}{z} =\frac{1}{\rho} e^{-i \pi / 4}=\frac{1}{\rho} e^{i {7\pi}/{4}},
$$
即三角形的斜边对应于 $w$ 平面内的射线 $|w| \geqslant \sqrt{2}/2, \arg w = 7\pi/4$.
- 直角边 $x \in[0,1], y=0$. 由
$$
u+i v=\frac{1}{z}=\frac{1}{x^2+y^2}(x-i y), 
$$
可得
\begin{equation}
u=\frac{x}{x^2+y^2},\quad v=-\frac{y}{x^2+y^2}. \label{eq:ex1_uv}
\end{equation}
代入 $y =0$ 与 $x \in[0,1]$, 可得 $ u = 1/x \geqslant 1, v = 0$, 这表示 $w$ 平面上 $u$ 轴上 $u\geqslant 1$ 的一条射线.
- 直角边 $x=1, y \in[0,1]$. 将 $x=1$ 代入 \eqref{eq:ex1_uv} 式, 可得参数方程
\begin{align}
u&=\frac{1}{1+y^2}, \label{eq:ex1_u} \\
v&=-\frac{y}{1+y^2}. \label{eq:ex1_v}
\end{align}
综合上面两式, 可得 $v = - u y$, 即 $y = -v/u$. 将之代入 \eqref{eq:ex1_u} 式, 可得
\begin{equation}
u^2+v^2=u . \label{eq1_uv2}
\end{equation}
这个方程表示 $w$ 平面上的圆心为 $(1/2, 0)$, 半径为 $1/2$ 的圆. 而 $y \in[0,1]$ 则意味着 $u \in\left[{1}/{2}, 1\right]$, $v \in\left[-{1}/{2}, 0\right]$. 因此, 与原像线段 $x=1, y \in[0,1]$ 对应的像集为 $w$ 平面上圆心为 $(1/2, 0)$, 半径为 $1/2$ 的圆上弧度在 $[3\pi/2, 2\pi]$ 之间的一段圆弧.

综上, $z$ 平面内的等腰直角三角形 $0\leqslant y \leqslant x, x\in [0,1]$ 在 $w$ 平面内对应的像集为射线 $u\geqslant 1, v=0$, 射线 $\arg(w) = 7\pi/4, |w| \geqslant \sqrt{2}/2$ 和圆心为 $(1/2, 0)$, 半径为 $1/2$ 的圆围成的辐角在 $[7\pi/4, 2\pi]$ 内的区域; 见下图.

此外, 我们还可以注意到, 上面例子中 $z$ 平面内曲线之间的夹角 (下图中的 $\alpha,\beta,\gamma$) 大小在经过 $w = 1/z$ 的映射之后, 与在 $w$ 平面的相应的像曲线之间的夹角 (下图中的 $\alpha',\beta',\gamma'$) 大小相同. 这是后面会讲到的保角变换或共形映射 (conformal mapping) 的一个例子.

**例 3:** 考虑 $z$ 平面内的等腰直角三角形: $0\leqslant y \leqslant x, x\in [0,1]$, 求其在 $w = \operatorname{Re}z + i \operatorname{Im} z/2$ 映射下 $w$ 平面内的像集.

**解:** 我们仍然分别考察三角形的三条边.

- 斜边. 此时, 我们利用 \eqref{eq:segment} 式将斜边参数化为 $z=(1+i) t$ $(t \in[0,1])$, 于是, 
$$
w = \left( 1 + \frac{i}{2} \right)t  \quad (t \in[0,1]),
$$
它表示 $w$ 平面内起点为原点, 终点为 $1+i/2$, 斜率为 $1/2$ 的线段.

- 直角边 $x \in[0,1], y=0$. 其 $w$ 平面内的像集是长度为 1 的线段, 参数方程为 $$w = x\quad (x\in [0,1]).$$

- 直角边 $x=1, y \in[0,1]$. 其 $w$ 平面内的像集是长度为 1/2 的线段, 参数方程为 $$w = 1 + i y/2\quad (y\in [0,1]).$$

从下图中可以看到, $w = \operatorname{Re}z + i \operatorname{Im} z/2$ 这一映射并不保持原像的曲线之间的夹角 ($\alpha''\neq \alpha, \gamma'' \neq \gamma$). 后面我们会明白这一映射与例 2 中的映射之间的本质区别.

#### 区域

在分析过上面的具体例子之后, 我们来给出一些基本概念的定义.

- 邻域: 以复数 $z_0$ 为圆心, 以任意小的正实数 $\varepsilon>0$ 为半径的圆内所有点 (即点集 $\{z  |\, |z-z_0|< \varepsilon\}$) 的集合称为点 $z_0$ 的 $\varepsilon$ 邻域 ($\varepsilon$-neighborhood). 不包括点 $z_0$ 的点集 $\{z |\, 0<|z-z_0|< \varepsilon\}$ 称为 $z_0$ 的 $\varepsilon$ 空心邻域 (deleted $\varepsilon$-neighborhood). 在不需要指明领域大小的情况下, 通常直接称为邻域和空心邻域. 当 $z_0= \infty$, 则称点集 $\{z |\, |z-a| > 1/\varepsilon\}$ 为 $\infty$ 的 $\varepsilon$ 邻域, 其中 $a$ 为任意有限复数; 由于 $\infty \pm a = \infty\, (a\neq \infty)$, $|z-a| > 1/\varepsilon$ 也可写为 $|z| > 1/\varepsilon$.

- 内点: 若点集 $E$ 包含 $z_0$ 的某个邻域, 即 $z_0$ 及其邻域内所有的点均属于 $E$, 则称 $z_0$ 为 $E$ 的内点 (interior point). 如果点集 $E$ 内所有的点都是其内点, 则称该点集为开集 (open set).

- 外点: 若 $z_0$ 及其某个邻域内所有的点均不属于点集 $E$,  则称 $z_0$ 为 $E$ 的外点 (exterior point).

- 边界点: 若 $z_0$ 的任一邻域内均有属于点集 $E$ 也有不属于 $E$ 的点, 则称 $z_0$ 为 $E$ 的边界点 (boundary point). 全体边界点的集合称为边界 (boundary). 如果点集 $E$ 包含其边界, 则称该点集为闭集 (closed set).

- 区域 (domain) 指满足以下两个条件的点集:
    - 全由内点组成, 即是开集;
    - 具有道路连通性, 即点集中的任意两点都可以用一条曲线连接起来, 并且这条曲线上所有的点都属于该点集.
    
    我们采取梁书的记号, 将区域记作 $B$ (德语单词 Bereich). 区域的重要性在后面讨论解析函数的时候会体现出来. 区域还可分为单连通 (singly-connected) 区域和多连通 (multiply-connected) 区域: 如果区域内任意一条简单闭曲线内部的点都属于该区域, 则称之为单连通区域; 否则称为多连通区域. 直观来看, 多连通区域便是带洞的区域.

- 闭区域: 区域 $B$ 内及其边界上所有点的集合构成闭区域 (closed domain), 记作 $\bar B$.

- 有界区域与无界区域: 如果存在一个有限半径的圆可以将区域里所有的点包围起来, 即 $\exists R >0$, 使得区域内所有的点 $z$ 都有 $|z| < R$, 则称该区域有界 (bounded); 否则, 则称之为无界 (unbounded) 区域.

#### 极限与连续

- 聚点与孤立点: 若 $z_0$ 的任一空心邻域内均有属于点集 $E$ 的点, 则称 $z_0$ 为 $E$ 的聚点 (accumulation point) 或极限点 (limit point). 若 $z_0\in E$, 但又不是 $E$ 的聚点, 也就是说, 有一个 $z_0$ 的空心邻域, 其内所有的点都不属于 $E$, 则称 $z_0$ 为 $E$ 的孤立点 (isolated point). 从上面的定义可知, 点集 $E$ 的聚点一定是其内点或者边界点. 

**例 4:** 令 $E_1$ 为区域 $|z|<1$, $z_1 = 1.2i$, 点集 $E = E_1 \cup \{z_1\} $, 则上述定义的各种点如下表及图所示: 

\begin{array}{l l}
\hline
内点 & |z|<1 \\
外点 & |z|>1,\text{且} z\neq z_1 \\
边界点 & |z|=1,\text{以及} z_1\\
聚点 & |z|\leqslant 1\\
孤立点 & z_1 \\\hline
\end{array}

这个例子中的集合既不是开集 (因为 $z_1$ 不是其内点), 也不是闭集 (因为不包含 $|z|=1$ 上的边界点).

- 极限: 给定一个点集 $E$, 对于 $z\in E$ 有函数 $w = f(z)$, 设 $z_0$ 为 $E$ 的一个聚点 (即内点或非孤立的边界点), 若存在一个复数 $w_0$, 对于任意的 $\varepsilon >0$, 总能找到 $\delta>0$, 使得当 $0 < |z -z_0| < \delta$ 时, 有
$$
\left|f(z)-w_0\right|<\varepsilon,
$$
则称$f(z)$ 沿 $E$ 在 $z_0$ 点有极限 (limit), 极限为 $w_0$,  记作
$$
\lim _{\substack{z \rightarrow z_0 \\ z\in E}} f(z)=w_0.
$$
其几何意义为: 存在 $z_0$ 的一个 $\delta$ 空心领域, 当自变量 $z$ 进入这个空心邻域时, 其像 $w$ 处于 $w_0$ 的一个任意小的 $\varepsilon$ 领域内. 为简单起见, 下面提到极限时, 只要意义明确, 就不再写沿着什么点集, 这时将极限记为
$$
\lim _{z \rightarrow z_0} f(z)=w_0.
$$

    注意以下两点: 
    1. $z_0$ 既可以是 $E$ 的内点, 也可以是 $E$ 的非孤立的边界点;
    2. $z$ 是复变量, 因此从 $z$ 到 $z_0$ 可以有无限多种路径, 只有当沿所有的路径给出的极限都是 $w_0$ 时, $f(z)$ 才在 $z_0$ 点有极限 $\lim _{z \rightarrow z_0} f(z)=w_0$.

- 连续: 函数 $f(z)$ 在点集 $E$ 内有定义, $z_0\in E$ 且是 $E$ 的聚点, 若
$$
\lim _{z \rightarrow z_0} f(z)= f(z_0),
$$
则称函数 $f(z)$ 在 $z_0$ 点连续 (continuous). 也就是说, 任给 $\varepsilon >0$, 总能找到 $\delta>0$, 使得当 $|z-z_0|<\delta$ 时, 有 $|f(z) - f(z_0)| <\varepsilon$. 如果 $f(z) $ 在点集 $E$ 中的每一点都连续, 则称 $f(z)$ 在 $E$ 中连续, 是 $E$ 中的连续函数.

**定理:** 函数 $f(z) = u(x,y) + i v(x,y)$ 在 $z_0 = x_0 + i y_0$ 处连续的充要条件是以 $(x,y)$ 为自变量的二元函数 $u(x,y)$ 和 $v(x,y)$ 在 $(x_0, y_0)$ 处连续, 即
$$
\lim_{\substack{x\to x_0\\ y \to y_0}} u(x,y) = u(x_0, y_0), \quad \lim_{\substack{x\to x_0\\ y \to y_0}} v(x,y) = v(x_0, y_0).
$$
**证:** 由复数的几何表示可得任一复数 $z$ 总是满足以下不等式
\begin{equation}
    |\operatorname{Re} z|\, (\text{或} |\operatorname{Im} z|) \leqslant |z| \leqslant |\operatorname{Re} z| + |\operatorname{Im} z|.
\end{equation}
于是, 
\begin{equation}
\begin{aligned}
  |u(x,y) - u(x_0,y_0)|\, (\text{或} |v(x,y) - v(x_0,y_0)|) &\leqslant |f(z) - f(z_0)| \\
  &\leqslant |u(x,y) - u(x_0,y_0)| + |v(x,y) - v(x_0,y_0)|.
\end{aligned} \label{eq:cont_ineq}
\end{equation}

由 \eqref{eq:cont_ineq} 式中的第一个不等式可得必要条件, 即 $\forall \varepsilon >0$, $\exists \delta>0$, 使得当 $|z - z_0|<\delta$ 时有 $|f(z) - f(z_0)| <\varepsilon$, 于是 
$$|u(x,y) - u(x_0,y_0)| < \varepsilon, \quad |v(x,y) - v(x_0,y_0)| < \varepsilon,$$
即 $u(x,y)$ 和 $v(x,y)$ 在 $(x_0, y_0)$ 处连续.

同样, 利用 \eqref{eq:cont_ineq} 式中的第二个不等式可得充分条件.

于是, 利用高等数学中的知识, 我们便知道如果两个复变函数在某点连续, 那么它们的和, 差, 积, 商 (作为分母的函数在该点不为 0) 以及这两个函数的复合函数在该点都连续.

**例 5:** $f(z)=z+z^*$ 在 $z=0$ 点是否连续?

**解:** 令 $z = x + i y$, 则
$$
\lim _{z \rightarrow 0} f(z)=\lim _{x \rightarrow 0} 2 x=0=f(0),
$$
因此, $f(z)$ 在 $z=0$ 点连续.

**例 6:** 如下定义的函数在 $z=0$ 点是否连续?
$$
f(z)= \begin{cases}\dfrac{z}{z^*}+\dfrac{z^*}{z}, & z \neq 0, \\ 1, & z=0.\end{cases} 
$$

**解:** 当 $z \neq 0$ 时, 令 $z=p e^{i \varphi}$, 则
$$
f(z)=e^{2 i \phi}+e^{-2 i \varphi}=\cos 2 \varphi.
$$
当 $z$ 沿实轴趋向于 $0$ 时, 有 $\cos(2\varphi) = 1$; 而当 $z$ 沿虚轴趋向于 $0$ 时, 有 $\cos(2\varphi) = -1$. 因此, $f(z)$ 在 $z=0$ 处没有极限, 故也不连续.

### 复变函数举例


#### 指数函数 (Exponential function)

\begin{equation}
{e}^z= {e}^{x+ {i} y}= {e}^x  {e}^{{i} y}= {e}^x(\cos y+ {i} \sin y).
\end{equation}

#### 三角函数 (Trigonometric function)

我们可以通过指数函数来定义三角函数, 其中正弦和余弦函数为:
\begin{equation}
\begin{aligned}
&\sin z=\frac{1}{2 {i}}\left({e}^{{i} z}- {e}^{-{i} z}\right), \\
&\cos z=\frac{1}{2}\left({e}^{{i} z}+ {e}^{-{i} z}\right).
\end{aligned}
\end{equation}

由于 $e^{iz}$ 有周期 $2\pi$, 复变的正弦和余弦函数也与实变时相同是周期函数, 周期为 $2\pi$. 易验证, 它们仍满足我们熟悉的等式:
$$ \sin^2 z + \cos^2 z = 1. $$
令 $z = x + iy$, 我们有
\begin{align}
\sin (x+i y) &=\frac{1}{2 i}\left(e^{i(x+i y)}-e^{-i(x+i y)}\right) \notag \\
&= \cosh y \sin x+i\sinh y \cos x, \\
\cos (x+i y) &= \cosh y \cos x - i\sinh y \sin x,
\end{align}
可以看到它们的取值范围不再限制在 $[-1,1]$ 之间, 并且它们的模可以大于 1:
\begin{equation}
\begin{aligned}
|\sin (z)| &=\frac{1}{2} \sqrt{2 \cosh (2 y)-2 \cos (2 x)}\,, \\
|\cos z| &=\frac{1}{2} \sqrt{2 \cosh (2 y)+2 \cos (2 x)}\,.
\end{aligned} \label{eq:sincosmod}
\end{equation}

#### 双曲函数 (Hyperbolic function)

我们也可以通过指数函数来定义复变的双曲函数, 其中双曲正弦和双曲余弦函数为:
\begin{equation}
\begin{aligned}
&\sinh z=\frac{1}{2}\left({e}^z-{e}^{-z}\right), \\
&\cosh z=\frac{1}{2}\left({e}^z+{e}^{-z}\right).
\end{aligned}
\end{equation}

可以看出, 它们与三角函数之间的关系为: 
\begin{equation}
\sinh (iz) = i \sin z, \quad \cosh (iz) = \cos z.
\end{equation}
易验证:
$$ \cosh^2 z - \sinh^2 z = 1. $$



#### 幂函数 (Power function)

\begin{equation}
z^n \quad (n \in \mathbb{N}, n >1).
\end{equation}


#### 多项式函数 (Polynomial function)

\begin{equation}
c_0+c_1 z+c_2 z^2+\ldots+c_n z^n=\sum_{i=0}^n c_i z^i  \quad (n \in \mathbb{N}),
\end{equation}
其中 $c_i$ 为多项式的系数, $n$ 为多项式的阶数.

#### 有理分式函数 (Rational function)

\begin{equation}
\frac{a_0+a_1 z+a_2 z^2+\ldots+a_n z^n}{b_0+b_1 z+b_2 z^2+\ldots+b_m z^m}=\frac{P(z)}{Q(z)}   \quad (m, n \in \mathbb{N}),
\end{equation}
其中 $P(z)$ 和 $Q(z)$ 分别为 $n$ 阶和 $m$ 阶的多项式函数.

#### 对数函数 (Logarithmic function)

\begin{equation}
\ln z=\ln \left(|z| {e}^{i\operatorname{Arg} z}\right)=\ln |z|+ i\operatorname{Arg} z,
\end{equation}



#### 一般幂函数 (General power function)

\begin{equation}
z^s= {e}^{s\ln z}.
\end{equation}
